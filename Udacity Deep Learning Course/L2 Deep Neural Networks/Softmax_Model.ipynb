# softmax 模型

下面编写的 softmax(x) 函数，应该返回一个形状和 x 相同的 NumPy array 类型

例如，当输入为一个列表或者一维矩阵（用列向量表示一个样本样本）时，比如说以下的：

```python
scores = [1.0, 2.0, 3.0]
```

应该返回一个同样长度（即3个元素）的一维矩阵：

```python
print softmax(scores)
[ 0.09003057  0.24472847  0.66524096]
```

对于一个二维矩阵，如以下（列向量表示单个样本），例如:

```python
scores = np.array([[1, 2, 3, 6],
                   [2, 4, 5, 6],
                   [3, 8, 7, 6]])
```

该函数应该返回一个同样大小(3,4)的二维矩阵，如以下:

```python
[[ 0.09003057  0.00242826  0.01587624  0.33333333]
 [ 0.24472847  0.01794253  0.11731043  0.33333333]
 [ 0.66524096  0.97962921  0.86681333  0.33333333]]
```

```python
"""Softmax."""

import numpy as np

scores = np.array([3.0, 1.0, 0.2])
scores1 = np.array([3.0, 1.0, 0.2]) * 10
scores2 = np.array([3.0, 1.0, 0.2]) / 10
print(scores)
print(scores1)
print(scores2)

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis = 0)

print('softmax(scores) = ',softmax(scores))
print('softmax(scores * 10) = ',softmax(scores1))
print('softmax(scores / 10) = ',softmax(scores2))

# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores).T, linewidth=2)
plt.show()
```

通过对比 softmax(scores)、softmax(scores * 10)、softmax(scores / 10)结果发现：

if you increase the size of your outputs，your classifier becomes very confident about its predictions.

But if you reduce the size of your outputs, your classifier becomes very unsure.

记住这一点供以后用，我们想让分类器在开始时不那么自信，然后随着时间的推移 它会因为学习增加信心。
